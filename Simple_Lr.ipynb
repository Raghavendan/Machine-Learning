```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ds=pd.read_excel('DS1.xls',engine="xlrd")
ds.head()
```

```python
ds['i']=1
```

```python
x=ds[['i','X']]
y=ds[['Y']]
```

```python
xt=x.T
print(f"Transpose of X{xt}")
```

```python
a=np.linalg.inv(xt@x)@xt@y
print(f"Intercept And Slope{a}")
```

```python
a.index=x.columns
pred=x@a
print("Prediction Value of Y's",pred)
```

```python
import matplotlib.pyplot as plt
```

```python
X=ds[['X']]
plt.scatter(X,y,alpha=0.5,marker="v")
plt.plot(X,pred,color="g")
plt.xlabel("Xaxis")
plt.ylabel("Yaxis")
plt.title("Linear Regression")
```
